```python
fp_canada=pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
fp_canada.head()# list the 5 head rows
```

```python
# Let's set Country as the index, it will help you to plot the charts easily, by refering to the country names as index value
fp_canada.set_index('Country',inplace=True)
# tip: The opposite of set is reset. So to reset the index, we can use df_can.reset_index()
```

```python
# useful for plotting later on
years = list(map(str, range(1980, 2014)))
years
```

```python
# Visualizing Data using Matplotlib
# Matplotlib.Pyplot
# One of the core aspects of Matplotlib is matplotlib.pyplot.
# It is Matplotlib's scripting layer which we studied in details in the videos about Matplotlib. 
# Recall that it is a collection of command style functions that make Matplotlib work like MATLAB. 
# Each pyplot function makes some change to a figure: e.g., creates a figure, creates a plotting area in a figure, 
# plots some lines in a plotting area, decorates the plot with labels, etc. 
# In this lab, we will work with the scripting layer to learn how to generate line plots. In future labs,
# we will get to work with the Artist layer as well to experiment first hand how it differs from the scripting layer.
```

```python
# What is a line plot and why use it?
# A line chart or line plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. 
# It is a basic type of chart common in many fields. Use line plot when you have a continuous data set.
# These are best suited for trend-based visualizations of data over a period of time.
```

```python
#creating data series
haiti = fp_canada.loc['Haiti', years] # passing in years 1980 - 2013 to exclude the 'total' column
haiti.head()
```

```python
haiti.plot()
```

```python
haiti.plot(kind='line')

#labeling y-axis and x-axis
plt.title("Immigration")
plt.ylabel('Numbers of immigration')
plt.xlabel('years')

plt.show()
```

```python
df_CI = fp_canada.loc[['India','China'], years]
df_CI.head()
```

```python
df_CI.plot(kind='line')
```

```python
df_CI=df_CI.transpose()
df_CI.head()
```

```python
# bar for vertical bar plots
# barh for horizontal bar plots
# hist for histogram
# box for boxplot
# kde or density for density plots
# area for area plots
# pie for pie plots
# scatter for scatter plots
# hexbin for hexbin plot
```

```python
df_CI.plot(kind ='line')

plt.title('Immigration')
plt.ylabel('No of Immigration')
plt.xlabel('Years')
plt.show()
```

```python

fp_canada.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = fp_canada.head(5)
df_top5 = df_top5[years].transpose()
print(df_top5)
 #Plot the dataframe. To make the plot more readeable, we will change the size using the `figsize` parameter.
df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top5.plot(kind='bar', figsize=(16, 8)) # pass a tuple (x, y) size
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')


plt.show()
```

```python
# Using line

fp_canada.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = fp_canada.head(5)
df_top5 = df_top5[years].transpose()
print(df_top5)
 #Plot the dataframe. To make the plot more readeable, we will change the size using the `figsize` parameter.
df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top5.plot(kind='line', figsize=(16, 8)) # pass a tuple (x, y) size
plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')


plt.show()
```
